# Extract numerical floor and total floors from the 'Floor' column

# We use a lambda function to handle different floor descriptions like 'Ground' or 'Lower Basement'

```python
HR['Floor Number'] = HR['Floor'].apply(lambda x: int(x.split(' out of ')[0]) if x.split(' ')[0].isdigit() else (0 if x.split(' ')[0].lower() == 'ground' else -1))
```

```python
HR['Total Floors'] = HR['Floor'].apply(lambda x: int(x.split(' out of ')[1]) if 'out of' in x else -1)
```

```python
def extract_floor(x):
    parts = x.split(' out of ')
    
    # Handle current floor
    if "Ground" in parts[0]:
        current = 0
    elif parts[0].strip().isdigit():
        current = int(parts[0].strip())
    else:
        current = np.nan

    # Handle total floors
    if len(parts) > 1 and parts[1].strip().isdigit():
        total = int(parts[1].strip())
    else:
        total = np.nan

    return pd.Series([current, total])
```

```python

HR[['Current Floor', 'Total Floors']] = HR['Floor'].apply(extract_floor)

```

```python
# Drop irrelevant columns
HR.drop(columns=['Floor', 'Area Locality', 'Point of Contact'], inplace=True)

```

```python
# Create new feature: Rent per sq.ft
HR['Rent_per_sqft'] = HR['Rent'] / HR['Size']

```
